```python
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('diabetes.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(df[['Glucose','BloodPressure','SkinThickness','Insulin','BMI','DiabetesPedigreeFunction','Age']],df['Outcome'],test_size=0.2,random_state=23)
```

```python
x_train_scaled=x_train.copy()
x_train_scaled['Age']=x_train_scaled['Age']/100
x_train_scaled['Glucose']=x_train_scaled['Glucose']/100
x_train_scaled['BloodPressure']=x_train_scaled['BloodPressure']/100
x_train_scaled['SkinThickness']=x_train_scaled['SkinThickness']/100
```

```python
x_test_scaled=x_test.copy()
x_test_scaled['Age']=x_test_scaled['Age']/100
x_test_scaled['Glucose']=x_test_scaled['Glucose']/100
x_test_scaled['BloodPressure']=x_test_scaled['BloodPressure']/100
x_test_scaled['SkinThickness']=x_test_scaled['SkinThickness']/100
```

```python
model=keras.Sequential(
    [keras.layers.Dense(100,activation='relu'),
     keras.layers.Dense(10,activation='relu'),
     keras.layers.Dense(1,activation='sigmoid')
])
```

```python
model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)

model.fit(x_train,y_train,epochs=500)
```

```python
model.evaluate(x_test,y_test)
```
